```python
@tf.function
def operations(x, y):
    return x + y, x - y, x * y, x / y

x = tf.random.uniform([3,2,2])
y = tf.random.uniform([3,2,2])

print(operations(x, y))
```

```python
#13
class SumModel(tf.Module):
    def __call__(self, x, y):
        return x + y

model = SumModel()
print(model(tf.constant(2), tf.constant(3)))
```

```python
class SubMulModel(tf.Module):
    def __call__(self, x, y):
        return x - y, x * y

model = SubMulModel()
print(model(tf.constant(5), tf.constant(2)))
```

```python
#15
class PerimeterModel(tf.Module):
    def __call__(self, a, b):
        return (a + b) * 2

model = PerimeterModel()
print(model(tf.constant(3), tf.constant(4)))
```

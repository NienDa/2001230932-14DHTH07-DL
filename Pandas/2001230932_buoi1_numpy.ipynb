```python
import numpy as np
import pandas as pd

# 1
print("Câu 1: ")
a = np.random.randint(0,21,(3,3))
print("C1:", a)
print(np.all(a != 0))
# 2
print("Câu 2: ")
a = np.random.randint(0,21,(3,3))
print("C2:", a)
print(np.any(a != 0))
# 3
print("Câu 3: ")
a = np.array([1,2,3])
b = np.array([2,2,1])
print("C3:")
print(np.greater(a,b))
print(np.greater_equal(a,b))
print(np.less(a,b))
print(np.less_equal(a,b))
# 4
print("Câu 4: ")
a = np.array([1]*10 + [0]*10 + [5]*10)
print("C4:", a)
# 5
print("Câu 5: ")
a = np.arange(30,71,2)
print("C5:", a)
# 6
print("Câu 6: ")
a = np.identity(3)
print("C6:\n", a)
# 7
print("Câu 7: ")
a = np.random.randint(15,56,10)
print("C7:", a[1:-1])

# 8
print("Câu 8: ")
a = np.random.randint(0,21,20)
a[(a>=9)&(a<=15)] *= -1
print("C8:", a)
# 9
print("Câu 9: ")
a = np.random.randint(10,22,(3,4))
print("C9:\n", a)
# 10
print("Câu 10: ")
a = np.ones((10,10))
a[1:-1,1:-1] = 0
print("C10:\n", a)
# 11
print("Câu 11: ")
a = np.diag([1,2,3,4,5])
print("C11:\n", a)
#
print("Câu 12: ")
a = np.random.randint(0,10,(3,3,3))
print("C12 sum row:\n", np.sum(a, axis=1))
print("C12 sum col:\n", np.sum(a, axis=2))

# 13
print("Câu 13: ")
a = np.random.randint(0,10,10)
b = np.random.randint(0,10,10)
print("C13:", np.dot(a,b))
# 14
print("Câu 14: ")
A = np.random.randint(0,10,(4,3))
y = np.random.randint(0,10,3)
print("C14:\n", A + y)

# 15
print("Câu 15: ")
s1 = pd.Series([2,4,6,8,10])
s2 = pd.Series([1,3,5,7,10])
print("C15:")
print(s1+s2)
print(s1-s2)
print(s1*s2)
print(s1/s2)

# 16
print("Câu 16: ")
df = pd.DataFrame({'A':[1,2,3],'B':[4,5,6]})
print("C16:\n", df.iloc[:,0])
# 17
print("Câu 17: ")
df = pd.DataFrame(np.random.rand(4,3), columns=['A','B','C'])
print("C17:\n", df.sort_values(by='A'))

# 18
print("Câu 18: ")
s = pd.Series([1,2,3], index=['a','b','c'])
print("C18:\n", s.reindex(['c','b','a']))
# 19
print("Câu 19: ")
sr1 = pd.Series([1,2,3,4,5])
sr2 = pd.Series([2,4,6,8,10])
print("C19:\n", sr1[~sr1.isin(sr2)])

# 20
print("Câu 20: ")
a = np.random.randint(0,50,20)
b = np.random.randint(0,50,20)
print("C20 union:", np.union1d(a,b))
print("C20 intersect:", np.intersect1d(a,b))

# 21
print("Câu 21: ")
x = pd.Series(range(10))
y = pd.Series(list("pqrstuvwxy"))
print("C21 vertical:\n", pd.concat([x,y]))
print("C21 horizontal:\n", pd.concat([x,y], axis=1))

# 22
print("Câu 22: ")
exam_data = {
'name': ['Anastasia','Dima','Katherine','James','Emily','Michael','Matthew','Laura','Kevin','Jonas'],
'score': [12.5,9,16.5,np.nan,9,20,14.5,np.nan,8,19],
'attempts': [1,3,2,3,2,3,1,1,2,1],
'qualify': ['yes','no','yes','no','no','yes','yes','no','no','yes']
}
labels = list('abcdefghij')
df = pd.DataFrame(exam_data, index=labels)
print("C22:\n", df)

# 23
print("Câu 23: ")
print("C23:\n", df.head(3))

# 24
print("C24:\n", df[['name','score']])

# 25
print("C25:\n", df[df['attempts'] > 2])

# 26
print("C26:", df.shape)

# 27
print("C27:\n", df[df['score'].isnull()])

# 28
print("C28:\n", df[df['score'].between(15,20)])

# 29
print("C29:\n", df[(df['attempts']>2) & (df['score'].between(15,20))])

# 30
df.loc['d','score'] = 19
print("C30:\n", df)
```
